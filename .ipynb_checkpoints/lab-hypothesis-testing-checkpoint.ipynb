In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

```python
df.info()
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
df['Type 2'] = df['Type 2'].fillna('')
df['Type 1'] = df['Type 1'].fillna('')
#code here
df_cleaned = df[['Name','Type 1','Type 2','HP']]
# Combine 'Type 1' and 'Type 2' columns
df_cleaned['Combined Type'] = df_cleaned['Type 1'] + '/' + df_cleaned['Type 2']
# Remove the trailing '/' if 'Type 2' was empty
df_cleaned['Combined Type'] = df_cleaned['Combined Type'].str.rstrip('/')
df_cleaned
```

```python
df_dragon = df_cleaned.loc[df_cleaned["Combined Type"].str.contains("Dragon"), "HP"]
df_grass = df_cleaned.loc[df_cleaned["Combined Type"].str.contains("Grass"), "HP"]
df_dragon
```

```python
t_stat, p_value = st.ttest_ind(df_dragon, df_grass, alternative='greater')
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
if p_value < 0.05:
    print("We reject the null hypothesis. There is significant evidence that Dragon-type Pokémon have more HP than Grass-type Pokémon.")
else:
    print("We fail to reject the null hypothesis. There is not enough evidence to suggest that Dragon-type Pokémon have more HP than Grass-type Pokémon.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df_legend = df.loc[df["Legendary"] == True]
df_nonlegend = df.loc[df["Legendary"] == False]
df_legend_cleaned= df_legend[['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']]
df_nonlegend_cleaned= df_nonlegend[['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']]
# Get list of all of stats : HP	Attack	Defense	Sp. Atk	Sp. Def	Speed	
# Multiple list  pull out hp 
```

```python
f_stat, fp_value =st.f_oneway(df_legend_cleaned, df_nonlegend_cleaned)
print(f"t-statistic: {f_stat}")
print(f"p-value: {fp_value}")
```
